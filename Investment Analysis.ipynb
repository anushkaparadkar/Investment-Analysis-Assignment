```python
#Removing the special characters in Permalink column.
companies["permalink"] = companies["permalink"].str.encode('ascii','ignore').str.decode('ascii').str.lower()
```

## Checkpoint 4 : Sector Analysis - 1

```python
#Importing mapping csv
mapping=pd.read_csv('data/mapping.csv',engine='python')
```

```python
#Extract the primary sector of each category list from the category_list column
master_frame['Primary_Sector']=master_frame.category_list.str.split('|').str[0].str.lower()
```

```python
# we have found our top 3 countries ,hence filtering out other values
master_frame = master_frame[master_frame['country_code'].isin(['USA', 'GBR', 'IND'])]
```

```python
master_frame.head()
```

```python
mapping.head(30)
```

```python
# cleaning mapping frame 
# removing NaN row
mapping = mapping[~mapping['category_list'].isna()]
```

```python
# converting all categories to lowercase
mapping['category_list'] = mapping['category_list'].str.lower()
mapping['category_list']
```

```python
#there are some categories which contain 0 in them
mapping[mapping['category_list'].str.contains('0')].head(15)
```

```python
#Replace all 0 with na and 
mapping['category_list']=mapping.category_list.str.replace('0','na')
```

```python
#after this step enterprise 2.0 becomes enterprise 2.na
mapping['category_list']=mapping.category_list.str.replace('2.na','2.0')
```

```python
# checking to see if any category with value ) is left
mapping[mapping['category_list'].str.contains('0')].head(15)
```

```python
melted=pd.melt(mapping, id_vars=["category_list"], var_name="Main_Sector",value_name="flag")
mapping_tag=melted[melted["flag"] == 1][['category_list','Main_Sector']]
mapping_tag
```

```python
main=pd.merge(master_frame,mapping_tag,left_on='Primary_Sector',right_on='category_list',how='inner')
```

```python
main.head()
```

```python
main.shape
```

```python
(main.groupby('Main_Sector')).Main_Sector.count()
```

```python
#Creating mapping dataframe and mapping each primary sector to one of the eight main sectors
mapping['category_list']= mapping['category_list'].replace({'0':'na','2.na':'2.0'}, regex=True).str.lower()
```

## Checkpoint 6: Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure()
```

```python
investment=plot_data.groupby('funding_round_type')['raised_amount_usd'].agg(['sum','mean']).reset_index()
investment=investment.rename(columns={'sum':'sum_invest','mean':'average_invest'})

investment['fraction']=100*(investment.sum_invest)/(investment.sum_invest.sum())
```

```python
investment['avg_usd_mil']=investment.average_invest/1000000
```

```python
investment
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=round(investment.avg_usd_mil),y=investment.fraction, hue=investment.funding_round_type,data=investment)
plt.title('Fund Type Analysis')
plt.show()
```

```python
# A plot showing the top 9 countries against the total amount of investments of funding type FT.
graph2=sns.barplot(x='country_code', y='raised_amount_usd', data=top9)
graph2.set(xlabel='Country Code', ylabel='Raised Amount in Million USD')
graph2.set_title('Country Analysis',fontsize =20)

plt.show()
```

```python
# A plot showing the number of investments in the top 3 sectors of the top 3 countries on one chart 

D1_top=D1[D1['Main_Sector'].isin(['Others','Social, Finance, Analytics, Advertising','Cleantech / Semiconductors'])]
D2_top=D2[D2['Main_Sector'].isin(['Others','Social, Finance, Analytics, Advertising','Cleantech / Semiconductors'])]
D3_top=D3[D3['Main_Sector'].isin(['Others','Social, Finance, Analytics, Advertising','Cleantech / Semiconductors'])]
```

```python
Dtotal=pd.concat([D1_top,D2_top,D3_top],axis=0)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='country_code', y='raised_amount_usd',hue = 'Main_Sector', data=Dtotal,estimator=lambda x: len(x))
plt.title('number of investments in the top 3 sectors of the top 3 countries')
plt.show()
```
