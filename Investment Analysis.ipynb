### Data Cleaning

```python
#Removing the special characters in Permalink column.
companies["permalink"] = companies["permalink"].str.encode('ascii','ignore').str.decode('ascii').str.lower()
```

```python
#Removing the special characters in name column.
companies["name"] = companies["name"].str.encode('ascii','ignore').str.decode('ascii').str.lower()
```

```python
companies.head()
```

```python
#Removing the special characters in the permalink column
rounds2["company_permalink"]=rounds2["company_permalink"].str.encode('ascii','ignore').str.decode('ascii').str.lower()
```

```python
#### Understanding the datasets
```

```python
#Unique comapnies in round2
rounds2["company_permalink"].nunique(dropna=True)
```

```python
#Unique companies in  companies.csv
companies["permalink"].nunique(dropna=True)
```

```python
#To check if company available in Companies dataframe is not avilable in rounds2 dataframe
companies.loc[~rounds2.company_permalink.isin(companies.permalink)]
```

```python
#Merge companies and rounds2 dataframes to a new dataframe named "master_frame"
master_frame=pd.merge(companies, rounds2, left_on='permalink', right_on='company_permalink', how='inner')
```

```python
master_frame.shape
```

## Checkpoint 2 - Investment Type Analysis


```python
master_frame.head()
```

```python
#dropping company_permalink column since it's same as permalink column
master_frame=master_frame.drop(['company_permalink'],1)
```

```python
master_frame
```

```python
#calculating the sum of null values in each column
master_frame.isnull().sum()
```

```python
#calculating percent of missing values
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#dropping funding_round_code due to high percent of missing values and deleting other irrelevant columns
master_frame=master_frame.drop(['funding_round_code'],1)
master_frame=master_frame.drop(['founded_at'],1)
```

```python
#calculating average of the investment types
pd.options.display.float_format = '{:.2f}'.format
avg_by_funding_type = master_frame.groupby(['funding_round_type'])['raised_amount_usd'].mean().sort_values() 
avg_by_funding_type
```

```python
#Suitable investment type for spark fund (between 5 to 15 million USD per investment round)
master_frame.groupby('funding_round_type')['raised_amount_usd'].mean().between(5000000,15000000,inclusive=True)
```

## Checkpoint 3 - Country Analysis 

```python
# Now that we know the suitable investment type,we can filter out all other types
master_frame = master_frame[(master_frame['funding_round_type'] == 'venture')]
```

```python
master_frame
```

```python
master_frame.shape
```

```python
# Grouping by countries to find the top9 country with the highest amount of funding
top9 = master_frame.groupby('country_code',as_index=False)['raised_amount_usd'].sum().sort_values(by = ['raised_amount_usd'],ascending=False).head(9)
```

```python
top9['Ranking'] = range(1,10)
top9.set_index(['Ranking'],inplace = True)
top9
```
